```python
#Removing other ( week days) then date letters using str[] fun
date=data['Date'].str[8:]
date.head()


#Removing Date Column Values 
data=data.drop('Date', axis=1)

#Concating Date Column with data set
df=pd.concat( [date, data], axis=1)
#df.head()
```

```python
plt.figure(figsize=(12,10))
df['Total Occ'].head(60).value_counts().head(10).plot.pie(autopct="%1.1f%%")
plt.show()
```

```python
#Total Occ more then 120 rooms 
group_A=df[df['Total Occ']>=120]
High= group_A['Total Occ']

#Total Occ Between 80 to 120 rooms
group_B=df[df['Total Occ'].between(80,120)]
Medium= group_B['Total Occ']

#Total Occ Between 60 to 80 rooms
group_C=df[df['Total Occ'].between(60,80)]
Low= group_C['Total Occ']

#Total Occ lessthan 60 Rooms
group_D=df[df['Total Occ']<=60]
Very_low= group_D['Total Occ']

print(High.size)
print(Medium.size)
print(Low.size)
print(Very_low.size)
```

```python
#Creating Data
labels= 'High', 'Medium', 'Low', 'Very_low'
sizes= [315, 183, 66, 21]

#colors= ['yellowgreen', 'glod', 'lightskyblue', 'lightcoral']
explode=(0, 0, 0, 0)

plt.pie(sizes, explode=explode, labels= labels,  autopct='%1.1f%%', shadow=True)

centre_circle= plt.Circle((0,0), 0.75, color='black', fc='white', linewidth=1.25)

fig= plt.gcf()
fig.gca().add_artist(centre_circle)

plt.axis('equal')
plt.show()
```

- 54% days  Total Occ is High (more then 120 Rooms)
- 3.6% days Total Occ is Very_Low (less than 50 Rooms)
